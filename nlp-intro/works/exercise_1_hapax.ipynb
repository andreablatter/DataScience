## Ejercicio 1 - Hapax

Un hapax legomenon es un termino que proviene del griego, y hace referencia a una palabra que ocurre una única vez dentro de un contexto. Este contexto, queda definido muchas veces por el caso de estudio.

---

El objetivo de este ejercicio, es encontrar el Hapax dentro del corpus del Quijote de la Mancha. Deberán es aplicar algunas técnicas de limpieza de datos, y tokenización para poder ver si existe o no dentrodel Quijote, un hapax.

Como se dijo, hapax puede depender de contexto que se defina. Para este ejercicio, el contexto será:
* El corpus completo.
* Una unidad a elección del grupo.

```python
# Imports necesarios.
import re
import os
import sys
import string
import collections
current_path = os.getcwd()
print (current_path)
```

```python
all_quijote=[]
with open("data/don_quijote.txt", "r") as f:
    for line in f:
        all_quijote.append(line) 
```

```python
len(all_quijote)
```

```python
all_quijote[0].strip()
```

```python
all_quijote=[x.strip() for x in all_quijote]
```

```python
new_quijote=[re.sub(r'\n','',x) for x in all_quijote]
```

```python
new_quijote
```

```python
#put all text in lower case
new_quijote=[x.lower() for x in all_quijote]
```

```python
#Create Default dictionary. Key is word and value 
word_count = collections.defaultdict(int)
print ("Debug 3- Collection Default Dictionary:", word_count)
```

```python
#- Split text and update result dictionary.
for i in new_quijote:
        for word in i.split():
            if word:#- Ignore whitespace.
                word_count[word] += 1
```

```python
 print ("Debug 4- Word and its count:", word_count)
```

```python
#- List which save word which value is 1.
hapax_words = list()
   for word, value in word_count.items():
        if value==1:
            hapax_words.append(word)
```

```python
print ("Debug 5- Final Hapax words:", hapax_words)
```

```python
print ("Count of Hapax words:", len(hapax_words))
```
